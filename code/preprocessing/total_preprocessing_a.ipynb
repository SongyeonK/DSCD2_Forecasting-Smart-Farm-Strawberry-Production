### type A 최종 데이터셋 구축 (환경 + 생육)

```python
import pandas as pd
```

#### 환경

```python
env_df = pd.read_csv("./2020/환경데이터_최최종_주별_a.csv", encoding="ANSI")
env_df.info()
env_df.head()
```

52번 임시삭제

```python
idx = env_df[env_df["id"] == 52].index

env_df.drop(idx, inplace=True)
```

```python
env_df.tail()
```

```python
# 4 ~ 30주차만 가져오기
env_df = env_df[(env_df["week"] > 3) & (env_df["week"] < 31)]
```

```python
env_df.columns
```

```python
# merge 용 컬럼 영문명으로 변경
env_df.rename(columns = {"s_am_max" : "sr_ex_max", "s_am_min" : "sr_ex_min", "s_am_mean" : "sr_ex_mean",
                        "st_max" : "soil_temp_max", "st_min" : "soil_temp_min", "st_mean" : "soil_temp_mean"}, inplace = True)
```

```python
env_df.info()
```

#### 생육 데이터

```python
grow_df = pd.read_csv("./2020/생육데이터_스무딩_완료.csv", encoding="ANSI")
grow_df.info()
grow_df.head()
```

```python
# # 사용할 컬럼만
# grow_col = ["농가명", "week", "개체번호", "초장", "엽수", "엽장", "엽폭", "엽병장", "관부직경", "화방별착과수"]

# grow_df = grow_df[grow_col]
```

```python
# grow_df = grow_df[(grow_df["week"] > 3) & (grow_df["week"] < 31)]
```

```python
# merge 용 컬럼 영문명으로 변경
grow_df.rename(columns = {"농가명" : "id", "개체번호" : "obj_num", "초장" : "height_pl", "엽수" : "num_leaf", "엽장" : "len_leaf", "엽폭" : "width_leaf", "엽병장" : "unit_len", "관부직경" : "dim_pipe", "화방별착과수" : "flw_get"}, inplace = True)
```

#### Merge 진행

merge 전 확인

```python
grow_df = grow_df[['id', 'week', 'obj_num', 'flw_get', 'last_flw_get', 'height_pl', 'num_leaf', 'len_leaf', 'width_leaf', 'unit_len', 'dim_pipe']]
```

```python
grow_df.info()
grow_df.head()
```

```python
env_df.info()
env_df.head()
```

생육 + 환경 join

```python
# 생육, 환경 join
fin_df = pd.merge(left=grow_df, right=env_df, how="left", on=["id", "week"])
```

```python
fin_df.columns
```

a 용

```python
cols = ['id', 'week', 'week_start_date', 'obj_num', 'flw_get', 
       'height_pl', 'num_leaf', 'len_leaf', 'width_leaf', 'unit_len', 'dim_pipe', 
       'temp_ex_max', 'temp_ex_min', 'temp_ex_mean', 
       'sr_ex_max', 'sr_ex_min', 'sr_ex_mean',
       'temp_int_max', 'temp_int_min', 'temp_int_mean',
       'hum_int_max', 'hum_int_min', 'hum_int_mean', 'CO2_max', 'CO2_min',
       'CO2_mean', 'soil_temp_max', 'soil_temp_min', 'soil_temp_mean']
fin_df = fin_df[cols]
```

```python
fin_df.to_csv("./2020/최최종데이터_a.csv", encoding="ANSI", index=False)
```

### 최종데이터에 전날 착과수 추가하는 코드

```python
import pandas as pd
```

```python
df = pd.read_csv("./2020/최최종데이터_a.csv")
df.info()
df.head()
```

```python
df.tail(8)
```

```python
# 정렬
df.sort_values(["id", "week", "obj_num"], inplace=True)
```

```python
last_flw_get = []

for f in df["id"].unique() :
    # print(f)
    last_flw_get += [None] * 4
    for w in range(5, 31) :
        # print(w)
        last_flw_get += df[(df["id"]==f) & (df["week"]==(w-1))]["flw_get"].tolist()

df["last_flw_get"] = last_flw_get
```

```python
df = df[['id', 'week', 'week_start_date', 'obj_num', 'flw_get', 'last_flw_get', 
       'height_pl', 'num_leaf', 'len_leaf', 'width_leaf', 'unit_len', 'dim_pipe',
       'temp_ex_max', 'temp_ex_min', 'temp_ex_mean', 'sr_ex_max', 'sr_ex_min',
       'sr_ex_mean', 'temp_int_max', 'temp_int_min', 'temp_int_mean',
       'hum_int_max', 'hum_int_min', 'hum_int_mean', 'CO2_max', 'CO2_min',
       'CO2_mean', 'soil_temp_max', 'soil_temp_min', 'soil_temp_mean']]
```

### 최종데이터에 푸리에 변환으로 week 변수 추가

```python
import pandas as pd

df = pd.read_csv("./2020/최최종데이터_a_지난주추가.csv")
df.info()
df.head()
```

```python
import numpy as np

# Fourier Features
df['week_sin'] = np.sin(2 * np.pi * df['week']/26.0)
df['week_cos'] = np.cos(2 * np.pi * df['week']/26.0)
```

```python
df = df[['id', 'week', 'week_start_date', 'obj_num', 'flw_get', 'week_sin', 'week_cos', 'last_flw_get', 
       'height_pl', 'num_leaf', 'len_leaf', 'width_leaf', 'unit_len', 'dim_pipe',
       'temp_ex_max', 'temp_ex_min', 'temp_ex_mean', 'sr_ex_max', 'sr_ex_min',
       'sr_ex_mean', 'temp_int_max', 'temp_int_min', 'temp_int_mean',
       'hum_int_max', 'hum_int_min', 'hum_int_mean', 'CO2_max', 'CO2_min',
       'CO2_mean', 'soil_temp_max', 'soil_temp_min', 'soil_temp_mean']]
```

```python
df.to_csv("./2020/최최종데이터_a_지난주추가_푸리에추가.csv", index=False)
```
