### 누락된 시간대 확인

```python
df = pd.read_csv("2020/1.환경/2020_환경_딸기_1차_농가_선정_시간대 처리.csv", encoding="ANSI")
```

```python
identity = pd.read_csv("2020/4.재배정보/2020_재배정보_딸기_1차_농가_선정.csv", encoding="ANSI")
```

```python
identity
```

```python
identity["정식일"] = pd.to_datetime(identity["정식일"])
df["측정시간"] = pd.to_datetime(df["측정시간"])
```

정식일부터 측정종료일까지

```python
farms = [2, 4, 5, 11, 16, 17, 20, 23, 24, 25, 26, 31, 34, 38, 49, 51, 52]
```

```python
farms = [2, 4, 5, 11, 16, 17, 20, 23, 24, 25, 26, 31, 34, 38, 49, 51, 52]

fin_df = pd.DataFrame()

for f in farms :
    print(f)

    # 비교할 시간대 설정
    sd = identity[identity["농가명"] == f]["정식일"].tolist()[0]
    ed = df[df["농가명"] == f]["측정시간"].tolist()[-1]

    print(sd, ed)
    # 실제로 있어야하는 시간대가 있는 데이터프레임
    rt = pd.DataFrame(pd.date_range(start=sd, end=ed, freq="h"), columns=["realTime"])
    rt["농가명"] = f

    sub_df = df[df["농가명"] == f]
    merge_df = rt.merge(sub_df, how='left', left_on=["realTime", "농가명"], right_on=["측정시간", "농가명"])

    fin_df = pd.concat([fin_df, merge_df])

    print(merge_df[merge_df.측정시간.isna()]["realTime"].tolist())

fin_df.to_csv("2020/1.환경/2020_환경_딸기_1차_농가_선정_누락 시간대 포함.csv", encoding="ANSI")
    
```
